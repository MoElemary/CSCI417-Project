```python
batch_size = 64

input_dim = 86

units = 330
output_size = 3  


def Neural_model(allow_cudnn_kernel=True):
    if allow_cudnn_kernel:
        
        long_short_term_memory_layer = keras.layers.LSTM(units, input_shape=(input_dim,1))
    else:
        
        long_short_term_memory_layer = keras.layers.RNN(
            keras.layers.LSTMCell(units), input_shape=(input_dim,1)
        )
    model = keras.models.Sequential(
        [
            long_short_term_memory_layer,
            keras.layers.BatchNormalization(),
            keras.layers.Dense(output_size),
        ]
    )
    return model
```

# New section

```python
 X = features
 y = label

 from sklearn.model_selection import train_test_split

 X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state = 42)
```

```python
Train_model = Neural_model(allow_cudnn_kernel=True)

Train_model.compile(
    loss=keras.losses.CategoricalCrossentropy(from_logits=True),
    optimizer="Adam",
    metrics=["categorical_accuracy"],
)

Train_model.summary()
```

```python
from tensorflow.keras.models import Model
```

```python
Train_model.fit(
    X_train, y_train, validation_data=(X_test, y_test), batch_size=batch_size, epochs=10
)
```
